```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/car_price_dataset.csv")
df
```

```python
 #1. What is the average price of all cars in the dataset?
#ANS:-
df['Price'].mean()
print(f"The average price of all cars in the dataset is ${round(df['Price'].mean(),2)}")
```

```python
#2. Which car brand has the highest average price?
#ANS:-
Brand=df.groupby("Brand")["Price"].mean().idxmax()
print(f"this is the car brand with highest average price is {Brand}")
```

```python
#3. What is the most common fuel type in the dataset?
#ANS:-
fuel=(df["Fuel_Type"].mode()[0])
print(f"this is the most common fuel type in the dataset is {fuel}")
```

```python
# 4. How many cars have a mileage greater than 200,000?
#ANS:-
len=(df[df["Mileage"] > 200000].shape[0])
print(f"Number of cars greater average than 200000 is: {len}")
```

```python
# 5. What is the distribution of cars by transmission type?
#ANS:-
Transmission = df["Transmission"].value_counts()
print(f"distribution of cars by transmission type is:\n{Transmission}")
```

```python
# 6. Which car model has the highest price?
#ANS:-
car = df.loc[df["Price"].idxmax(),"Model"]
print(f"this is the car model with highest price is {car}")
```

```python
# 7. What is the average price of cars by fuel type?
#ANS:-
average_price=df.groupby("Fuel_Type")["Price"].mean()
print(f"average price of cars by fuel type is:\n{average_price}")
```

```python
# 8. How many cars are from the year 2020 or later?
#ANS:-
cars= df[df['Year'] >= 2020].shape[0]
print(f"this is the number of cars are from the year 2020 or later is {cars}")
```

```python
# 9. What is the average engine size for each brand?
#ANS:-
average_engine= [df.groupby("Brand")["Engine_Size"].mean()]
print(f"average engine size for each brand is:\n{average_engine}")
```

```python
# 10. Which car has the lowest mileage?
#ANS:-
lowest_mileage=df.loc[df["Mileage"].idxmin(),"Model"]
mileage=df["Mileage"].min()
print(f"this is the car with lowest mileage is {lowest_mileage}")
print(f"this is the mileage of car is {mileage}")
```

```python
# 11. What is the correlation between mileage and price?
#ANS:-
correlation = df['Mileage'].corr(df['Price'])
print(f"this is the correlation between mileage and price is {correlation}")
```

```python
# 12. How many cars have more than 2 owners?
#ANS:-
len = df[df['Owner_Count'] > 2].shape[0]
print(f"this is the number of cars have more than 2 owners is {len}")
```

```python
# 13. What is the average price of cars by year?
#ANS:-
average_price= df.groupby('Year')['Price'].mean()
print(f"average price of cars by year is:\n{average_price}")
```

```python
# 14. Which car has the highest number of doors?
#ANS:-
df[df['Doors'] == df['Doors'].max()]
```

```python
# 15. What is the average mileage for each fuel type?
#ANS:-
average_mileage = df.groupby("Fuel_Type")["Mileage"].mean()
print(f"The average mileage for each fuel type is:\n{average_mileage}")
```

```python
# 16. How many cars are from the brand 'Toyota'?
#ANS:-
cars= df[df["Brand"] == "Toyota"].shape[0]
print(f"this is the number of cars are from the brand 'Toyota' is {cars}")
```

```python
# 17. What is the average price of cars with automatic transmission?
#ANS:-
average_price = df[df["Transmission"] == "Automatic"]["Price"].mean()
print(f"average price of cars with automatic transmission is ${round(average_price,2)}")
```

```python
# 18. Which car has the highest owner count?
#ANS:-
car=df.loc[df["Owner_Count"].idxmax(),"Model"]
owner=df["Owner_Count"].max()
print(f"this is the car with highest owner count is {car}")
print(f"this is the owner count of car is {owner}")
```

```python
# 19. What is the average engine size for cars with diesel fuel type?
#ANS:-
average_engine=df[df["Fuel_Type"] == "Diesel"]["Engine_Size"].mean()
print(f"average engine size for cars with diesel fuel type is {round(average_engine,2)}")
```
